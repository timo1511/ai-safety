# AI Shield Kenya Case Studies

This notebook demonstrates real-world applications of AI Shield Kenya in Kenyan agriculture and education contexts. It shows how the toolkit can be used to detect harmful content in practical scenarios faced by smallholder farmers and CBC students.

## Setup

Import the necessary modules:

```python
import sys
import os

# Add the src directory to Python path
sys.path.insert(0, os.path.join('..', 'src'))

# Import main modules
from ai_shield_kenya import RiskScorer
from ai_shield_kenya.detectors import FraudDetector, MisinformationDetector, ToxicityDetector
from ai_shield_kenya.knowledge_base import KenyanKnowledgeBase

print("Case studies modules imported successfully!")
```

# Agriculture Case Study

## Scenario: Smallholder Farmer Receives WhatsApp Message

A smallholder farmer in western Kenya receives a WhatsApp message offering agricultural advice. Let's analyze it with AI Shield Kenya:

```python
# WhatsApp message from unknown sender
whatsapp_message = "Ugonjwa wa MLN unaweza kutibiwa kwa kutumia bleach na asali ya mafuta ya neem. Tuma KSh 5000 kwa namba ya simu 0712345678 kufanya malipo ya dawa."

print(f"WhatsApp message:")
print(f"'{whatsapp_message}'")

# Analyze with RiskScorer
scorer = RiskScorer()
result = scorer.score(whatsapp_message, language="sw", domain="agriculture")

print(f"
Risk analysis result:")
print(f"Risk score: {result['risk_score']}")
print(f"Category: {result['category']}")
print(f"Suggestion: {result['suggestion']}")

# Detailed analysis with specialized detectors
fraud_detector = FraudDetector()
misinfo_detector = MisinformationDetector()
toxicity_detector = ToxicityDetector()

fraud_result = fraud_detector.detect(whatsapp_message, language="sw")
misinfo_result = misinfo_detector.detect(whatsapp_message, language="sw", domain="agriculture")
toxicity_result = toxicity_detector.detect(whatsapp_message, language="sw")

print(f"
Detailed detector analysis:")
print(f"Fraud detection: {fraud_result['detected']} (severity: {fraud_result['severity']})")
print(f"Misinformation detection: {misinfo_result['detected']} (severity: {misinfo_result['severity']})")
print(f"Toxicity detection: {toxicity_result['detected']} (severity: {toxicity_result['severity']})")
```

## Knowledge Base Verification

Verify claims against the Kenyan knowledge base:

```python
# Initialize knowledge base
kb = KenyanKnowledgeBase()

# Verify each claim
claims = [
    "Ugonjwa wa MLN unaweza kutibiwa kwa kutumia bleach",
    "asali ya mafuta ya neem",
    "Tuma KSh 5000 kwa namba ya simu 0712345678 kufanya malipo ya dawa"
]

print(f"Knowledge base verification:")
for claim in claims:
    verification = kb.verify_claim(claim)
    print(f"- '{claim[:30]}...' -> Verified: {verification['verified']}, Confidence: {verification['confidence']}, Source: {verification['source']}")
```

# SIR Model Visualization

Visualize how the SIR model works for different types of harm:

```python
import numpy as np
import matplotlib.pyplot as plt
from ai_shield_kenya.sir_model import SIRHarmModel

# Create SIR model instances for different harm types
# Fraud model
fraud_model = SIRHarmModel(beta=0.85, gamma=0.15)

# Misinformation model
misinfo_model = SIRHarmModel(beta=0.65, gamma=0.28)

# Toxicity model
toxicity_model = SIRHarmModel(beta=0.92, gamma=0.18)

# Simulate trajectories
def simulate_trajectory(model, N=100):
    t = np.linspace(0, 10, 100)
    initial_state = [N * 0.8, N * 0.1, N * 0.1]  # S, I, R
    try:
        solution = model._solve_sir(initial_state, t)
        return t, solution[:, 0], solution[:, 1], solution[:, 2]
    except:
        # Fallback for numerical issues
        return t, [N*0.8]*100, [N*0.1]*100, [N*0.1]*100

# Plot trajectories
plt.figure(figsize=(12, 8))

# Fraud trajectory
t_fraud, s_fraud, i_fraud, r_fraud = simulate_trajectory(fraud_model)
plt.subplot(2, 2, 1)
plt.plot(t_fraud, s_fraud, 'b-', label='Susceptible')
plt.plot(t_fraud, i_fraud, 'r-', label='Infected')
plt.plot(t_fraud, r_fraud, 'g-', label='Recovered')
plt.title('Fraud Harm Propagation')
plt.legend()
plt.grid(True, alpha=0.3)

# Misinformation trajectory
t_misinfo, s_misinfo, i_misinfo, r_misinfo = simulate_trajectory(misinfo_model)
plt.subplot(2, 2, 2)
plt.plot(t_misinfo, s_misinfo, 'b-', label='Susceptible')
plt.plot(t_misinfo, i_misinfo, 'r-', label='Infected')
plt.plot(t_misinfo, r_misinfo, 'g-', label='Recovered')
plt.title('Misinformation Propagation')
plt.legend()
plt.grid(True, alpha=0.3)

# Toxicity trajectory
t_toxicity, s_toxicity, i_toxicity, r_toxicity = simulate_trajectory(toxicity_model)
plt.subplot(2, 2, 3)
plt.plot(t_toxicity, s_toxicity, 'b-', label='Susceptible')
plt.plot(t_toxicity, i_toxicity, 'r-', label='Infected')
plt.plot(t_toxicity, r_toxicity, 'g-', label='Recovered')
plt.title('Toxicity Propagation')
plt.legend()
plt.grid(True, alpha=0.3)

# Combined comparison
plt.subplot(2, 2, 4)
plt.plot(t_fraud, i_fraud, 'r-', label='Fraud')
plt.plot(t_misinfo, i_misinfo, 'orange-', label='Misinformation')
plt.plot(t_toxicity, i_toxicity, 'purple-', label='Toxicity')
plt.title('Infection Comparison')
plt.ylabel('Infected Level')
plt.xlabel('Time')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"SIR model visualization complete. Each plot shows how harm propagates over time:")
print(f"- Blue: Susceptible content (not yet exposed)")
print(f"- Red: Infected content (actively harmful)")
print(f"- Green: Recovered content (corrected/safe)")
```
