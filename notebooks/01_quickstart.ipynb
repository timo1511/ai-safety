## Basic Usage

Create a RiskScorer instance and score some sample prompts:

````python
# Create risk scorer
scorer = RiskScorer()

# Test with English prompt
result_en = scorer.score("Use bleach to cure crop disease", language="en", domain="agriculture")
print(f"English prompt result:")
print(f"Risk score: {result_en['risk_score']}")
print(f"Category: {result_en['category']}")
print(f"Suggestion: {result_en['suggestion']}")

# Test with Swahili prompt
result_sw = scorer.score("Ugonjwa wa MLN unaweza kutibiwa kwa kutumia bleach", language="sw", domain="agriculture")
print(f"
Swahili prompt result:")
print(f"Risk score: {result_sw['risk_score']}")
print(f"Category: {result_sw['category']}")
print(f"Suggestion: {result_sw['suggestion']}")
````

## Using Specialized Detectors

You can also use specialized detectors for more detailed analysis:

````python
# Fraud detector
fraud_detector = FraudDetector()
fraud_result = fraud_detector.detect("Send KSh 5000 to M-Pesa number 0712345678 for fertilizer delivery", language="en")
print(f"Fraud detection result:")
print(f"Detected: {fraud_result['detected']}")
print(f"Severity: {fraud_result['severity']}")
print(f"Risk score: {fraud_result['risk_score']}")
print(f"Explanation: {fraud_result['explanation']}")

# Misinformation detector
misinfo_detector = MisinformationDetector()
misinfo_result = misinfo_detector.detect("CBC Math Grade 4: 2+2=5", language="en", domain="education")
print(f"
Misinformation detection result:")
print(f"Detected: {misinfo_result['detected']}")
print(f"Severity: {misinfo_result['severity']}")
print(f"Risk score: {misinfo_result['risk_score']}")
print(f"Explanation: {misinfo_result['explanation']}")
````

## Knowledge Base Queries

The knowledge base contains verified facts about Kenyan agriculture and education:

````python
# Initialize knowledge base
kb = KenyanKnowledgeBase()

# Query agriculture facts
agri_facts = kb.query("maize disease", language="en")
print(f"Agriculture facts about maize disease:")
for fact in agri_facts:
    print(f"- {fact['english']}")

# Verify a claim
verification = kb.verify_claim("Maize lethal necrosis (MLN) is a serious viral disease affecting maize in Kenya")
print(f"
Verification result:")
print(f"Verified: {verification['verified']}")
print(f"Confidence: {verification['confidence']}")
print(f"Source: {verification['source']}")
````

## Batch Processing

Process multiple prompts at once:

````python
# Sample prompts
prompts = [
    "Use bleach to kill aphids on tomatoes",
    "CBC Math Grade 4: 2+2=5",
    "Send KSh 5000 to M-Pesa number 0712345678 for fertilizer delivery",
    "The main planting season for maize in western Kenya is March to May"
]

# Score all prompts
results = []
for prompt in prompts:
    result = scorer.score(prompt, language="en", domain="agriculture")
    results.append(result)

# Display results
print(f"Batch processing results:")
for i, result in enumerate(results):
    print(f"{i+1}. '{prompts[i]}' -> Risk: {result['risk_score']}, Category: {result['category']}")
````

## Visualization

Plot risk scores for different prompts:

````python
import matplotlib.pyplot as plt
import numpy as np

# Extract risk scores
risk_scores = [r['risk_score'] for r in results]
prompt_labels = [p[:30] + '...' if len(p) > 30 else p for p in prompts]

# Create bar chart
plt.figure(figsize=(10, 6))
plt.bar(range(len(risk_scores)), risk_scores, color=['red' if s > 60 else 'orange' if s > 30 else 'green' for s in risk_scores])
plt.xticks(range(len(risk_scores)), prompt_labels, rotation=45, ha='right')
plt.ylabel('Risk Score (0-100)')
plt.title('Risk Scores for Sample Prompts')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"Risk score summary:")
print(f"Average risk: {np.mean(risk_scores):.1f}")
print(f"Max risk: {np.max(risk_scores):.1f}")
print(f"Min risk: {np.min(risk_scores):.1f}")
````
